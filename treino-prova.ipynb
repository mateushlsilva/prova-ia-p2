```python
from sklearn.cluster import KMeans 

import pandas as pd 

from sklearn.preprocessing import MinMaxScaler 

from matplotlib import pyplot as plt 
```

```python
df = pd.read_csv('Popular_Spotify_Songs.csv', encoding='latin-1')
```

```python
# GrÃ¡fico de DispersÃ£o: Spending Score (1-100) x Annual Income (K$) 

plt.scatter(df['in_apple_playlists'],df['in_spotify_playlists']) 

plt.xlabel("in_apple_playlists") 

plt.ylabel('in_spotify_playlists') 
```

```python
# WCSSE 

sse = [] 

k_rng = range(1,4) 

for k in k_rng: 

    km = KMeans(n_clusters=k) 

    km.fit(df[['in_apple_playlists','in_spotify_playlists']]) 

    sse.append(km.inertia_) 

print(sse) 
```

```python
plt.title('O Metodo Elbow') 

plt.xlabel('Qtde. de Clusters (K)') 

plt.ylabel('WCSSE - Sum of Squared Errors') # Within Cluster Sum of Squares 

plt.plot(k_rng,sse) 
```

```python
# ConstruÃ§Ã£o do Modelo de ML 

km = KMeans(n_clusters=3) #  
```

```python
# AplicaÃ§Ã£o do Modelo: â€¯ â€¯ â€¯Index: â€¯0 â€¯ â€¯ â€¯ 1 â€¯ â€¯ â€¯ â€¯ â€¯ 2 â€¯ â€¯ â€¯ â€¯ â€¯ â€¯ â€¯ â€¯ â€¯ â€¯ â€¯3 

y_predicted = km.fit_predict(df[['in_apple_playlists','in_spotify_playlists']]) 

y_predicted 
```

```python

km.cluster_centers_  
```

```python

# InclusÃ£o do CentrÃ³ide como uma nova coluna do Dataframe 

df['cluster']=y_predicted 

df.head() 
```

```python
km.cluster_centers_
```

```python
# ComparaÃ§Ã£o de Clusters: Spending Score (1-100) vs. Annual Income (k$) 

# Index nos CentrÃ³ides: 0-Gender; 1-Age; 2-Spending Score (1-100); 3-Annual Income (k$) 

df1 = df[df.cluster==0] 

df2 = df[df.cluster==1] 

df3 = df[df.cluster==2] 

# in_apple_playlists','in_spotify_playlists


plt.scatter(df1['in_apple_playlists'],df1['in_spotify_playlists'] ,color='green',label='Cluster 0') 

plt.scatter(df2['in_apple_playlists'],df2['in_spotify_playlists'] ,color='red',label='Cluster 1') 

plt.scatter(df3['in_apple_playlists'],df3['in_spotify_playlists'] ,color='black',label='Cluster 2') 


plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid') 

plt.xlabel("in_apple_playlists") 

plt.ylabel('in_spotify_playlists') 

plt.legend() 
```
